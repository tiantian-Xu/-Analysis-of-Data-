```python
import nltk
import jieba
import pandas as pd
```

```python
file_path = open('./商品评价信息.csv')
file_data = pd.read_csv(file_path)
file_data
```

```python
# 删除重复文件
file_data = file_data.drop_duplicates()
```

```python
# 使用精确模式划分中文
cut_words = jieba.lcut(str(file_data['评价信息'].values), cut_all = False)
```

```python
# 加载停用词表
file_path = open('./停用词表.txt',encoding='utf-8')
stop_words = file_path.read()
# 删除停用词
# 新建一个空列表，用于存储删除停用词后的数据
new_data = []
for word in cut_words:
    if word not in stop_words:
        new_data.append(word)
```

```python
# 词频统计
from nltk import FreqDist
freq_list = FreqDist(new_data)
# 返回词语列表
freq_list = freq_list.most_common()
freq_list
```

```python
from matplotlib import pyplot as plt
from wordcloud import WordCloud
# 词云显示
font = './FZYouHFW.TTF'           # 华文行楷
words = ",".join(new_data)
wc = WordCloud(font_path = font,                 #解决显示口字乱码问题
               background_color = 'white',        #背景颜色
               width = 1200, height =1000,     
               max_font_size = 250,
                max_words=30, 
               prefer_horizontal=15        #   调整词云中字体水平和垂直的多少
              ).generate(words)
plt.imshow(wc)   # 用plt显示图片
plt.axis('off')  # 不显示坐标轴
plt.savefig('商品评价词云.png')
plt.show()       # 显示图片
```
