```python
# Tushare是一个免费、开源的python财经数据接口包
import tushare as ts
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import datetime
today=str(datetime.date.today())
today
```

```python
# 获取深圳成指K线数据,向前复权
sz = ts.get_k_data('sz',start='2019-10-01', end = today, autype='qfq')
# 获取上证指数K线数据,向前复权
sh = ts.get_k_data('sh',start='2019-10-01', end = today, autype='qfq')
```

```python
# 获取事件的横坐标
# 2019/12/8武汉卫健委通报首例新冠状病例当日发病 - 移到交易日
d1 = sh[sh.date == '2019-12-10'].index.tolist()[0] # index = 228

# 2019/12/20稳健医药取消在湖北召开高管年终会议
d2 = sh[sh.date == '2019-12-20'].index.tolist()[0] # index = 236

# 2019/12/31国家卫健委来汉-武汉卫健委称:未见明显人传人和医护感染
d3 = sh[sh.date == '2019-12-31'].index.tolist()[0] # index = 243

# 2020/1/2香港特区政府开会检视针对武汉疫情的预防措施
d4 = sh[sh.date == '2020-01-02'].index.tolist()[0] 

# 香港将武汉不明肺炎列入传染病
d5 = sh[sh.date == '2020-01-08'].index.tolist()[0] 

# 2020/1/22湖北省启动突发公共事件二级响应，2020/1/23武汉封城
d6 = sh[sh.date == '2020-01-22'].index.tolist()[0]

# 2020/2/3央行开展1.2万亿元公开市场逆回购操作投放资金
d7 = sh[sh.date == '2020-02-03'].index.tolist()[0] 

# 2020/2/6中国消减对原产于美国约750美元进口商品加征关税
d8 = sh[sh.date == '2020-02-06'].index.tolist()[0]
```

```python
# 日期
date = sh['date']
# 收盘价
sz = sz['close']
sh = sh['close']
```

```python
import matplotlib.pyplot as plt
# 支持中文
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
```

```python
# 样式美化(plt.style.use)定制画布风格
plt.style.use('ggplot')
# 创建图形
plt.figure(figsize=(20,15))
# 创建两个子图中的第一个，设置坐标轴
# 上证指数K线
plt.subplot(2, 1, 1) # (行、列、子图编号)
plt.plot(sh,color="black")
plt.title('2019/10/1至今日 上证指数K线数据')

plt.vlines(d1, 2700, 3200, colors='k')
plt.text(d1-1,2800,'d1', ha='center', va= 'bottom',fontsize=12)

plt.vlines(d2, 2700, 3200, colors='k')
plt.text(d2-1,2800,'d2', ha='center', va= 'bottom',fontsize=12)

plt.vlines(d3, 2700, 3200, colors='k')
plt.text(d3-1,2800,'d3', ha='center', va= 'bottom',fontsize=12)

plt.vlines(d4, 2700, 3200, colors='k')
plt.text(d4+1,2800,'d4', ha='center', va= 'bottom',fontsize=12)

plt.vlines(d5, 2700, 3200, colors='k')
plt.text(d5-1,2800,'d5', ha='center', va= 'bottom',fontsize=12)

plt.vlines(d6, 2700, 3200, colors='k')
plt.text(d6-1,2800,'d6', ha='center', va= 'bottom',fontsize=12)

plt.vlines(d7, 2700, 3200, colors='k')
plt.text(d7-1,2800,'d7', ha='center', va= 'bottom',fontsize=12)

plt.vlines(d8, 2700, 3200, colors='k')
plt.text(d8-1,2780,'d8', ha='center', va= 'bottom',fontsize=12)

# 深圳成指K线数据
plt.subplot(2, 1, 2)
plt.plot(sz,color="black")
plt.title('2019/10/1至今日 深圳成指K线数据')

plt.vlines(d1, 9500, 11500, colors='k')
plt.text(d1-1,11200,'d1', ha='center', va= 'bottom',fontsize=12)

plt.vlines(d2, 9500, 11500, colors='k')
plt.text(d2-1,11200,'d2', ha='center', va= 'bottom',fontsize=12)

plt.vlines(d3, 9500, 11500, colors='k')
plt.text(d3-1,11200,'d3', ha='center', va= 'bottom',fontsize=12)

plt.vlines(d4, 9500, 11500, colors='k')
plt.text(d4+1,11200,'d4', ha='center', va= 'bottom',fontsize=12)

plt.vlines(d5, 9500, 11500, colors='k')
plt.text(d5-1,11200,'d5', ha='center', va= 'bottom',fontsize=12)

plt.vlines(d6, 9500, 11500, colors='k')
plt.text(d6-1,11200,'d6', ha='center', va= 'bottom',fontsize=12)

plt.vlines(d7, 9500, 11500, colors='k')
plt.text(d7-1,11200,'d7', ha='center', va= 'bottom',fontsize=12)

plt.vlines(d8, 9500, 11500, colors='k')
plt.text(d8-1,11200,'d8', ha='center', va= 'bottom',fontsize=12)

plt.savefig('./新冠状病毒对中国股市影响.png')
```
