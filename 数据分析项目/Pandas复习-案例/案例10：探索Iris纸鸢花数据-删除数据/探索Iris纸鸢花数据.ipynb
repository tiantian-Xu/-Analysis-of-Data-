```python
import pandas as pd 
import numpy as np
```

```python
iris = pd.read_csv('./iris.csv',names = ['sepal_length','sepal_width', 'petal_length', 'petal_width', 'class'] )
iris.head()
```

# 数据框中有缺失值吗？ *

```python
pd.isnull(iris).sum()
```

# 将列petal_length的第10到19行设置为缺失值

```python
iris['petal_length'][10:20] = np.nan
iris.head(20)
```

# 将缺失值全部替换为1.0

```python
iris.fillna(1,inplace=True)
```

- inplace：布尔型，默认为False。如果为True，则就地填充。注意：这将修改此对象上的所有其他视图

# 删除列class

```python
del iris['class']
iris.head()
```

# 将数据框前三行设置为缺失值

```python
iris.loc[0:2,] = np.nan
iris.head()
```

# 删除有缺失值的行

```python
iris = iris.dropna()
iris.head()
```

# 重新设置索引 *

```python
iris = iris.reset_index(drop = True) # drop = true 删除原来的索引
iris.head()
```
