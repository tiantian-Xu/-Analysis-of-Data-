```python
import pandas as pd
```

```python
drink = pd.read_csv('./drinks.csv')
```

```python
drink.head()
```

# 哪个大陆(continent)平均消耗的啤酒(beer)更多？

```python
beer = drink.groupby('continent')['beer_servings'].mean()
beer.sort_values(ascending=False)
```

# 打印出每个大陆(continent)的红酒消耗(wine_servings)的描述性统计值

```python
drink.groupby('continent')['wine_servings'].describe()
```

# 打印出每个大陆每种酒类别的消耗平均值

```python
round(drink.groupby('continent').mean(),2)
```

# 打印出每个大陆每种酒类别的消耗中位数

```python
round(drink.groupby('continent').median(),2)
```

# 打印出每个大陆对spirit饮品消耗的平均值，最大值和最小值

```python
drink.groupby('continent').spirit_servings.agg(['mean','max','min'])
```
