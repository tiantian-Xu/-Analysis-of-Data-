# 数据读取

```python
import pandas as pd
import numpy as np
file_path = open('D:/Date/链家北京租房数据.csv')
file_data = pd.read_csv(file_path)
file_data
```

# 数据预处理
## 重复值

```python
# 删除重复数据
file_data = file_data.drop_duplicates()
# file_data
```

## 缺失值

```python
# 删除缺失数据
file_data = file_data.dropna()
# file_data
```

## 数据转换类型

```python
# 删除“面积(㎡)”列中的中文字符
#创建空数组
data_S = np.array([])
# 获取面积数据
data = file_data['面积(㎡)'].values
# 遍历数据
for i in data:
    data_S = np.append(data_S, np.array(i[:-2]))
data = data_S.astype(np.float)
file_data.loc[:,'面积(㎡)'] = data
file_data
```

# 图表分析
## 房源数量、位置分布分析

```python
#根据区域对象分组
house_count = file_data.groupby(by=['区域']).count()['户型']
house_count
```

```python
import matplotlib.pyplot as plt
# 支持中文
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
# house_count.plot(kind='bar',figsize = (20,8))
house_count.sort_values(ascending=False).plot(kind='bar',figsize = (10,4))
# 设置标题
plt.title("房源数量&位置表")
plt.xlabel("房源位置")
plt.ylabel("房源数量")
# 保存
plt.savefig('./房源数量&位置表.png')
# 显示图例说明
plt.legend(loc='best')
# 展示
plt.show()
```

## 户型数量分析

```python
# 规范户型格式，将“房间”替换成“室”
housetype_data = file_data['户型']
temp_list = []
# 通过replace()方法进行替换
for i in housetype_data:
    new_info = i.replace('房间','室')
    temp_list.append(new_info)
file_data.loc[:,'户型'] = temp_list
file_data
```

```python
housetype_count = file_data.groupby(by=['户型']).count()['区域']
# housetype_count
```

```python
#户型种类数
housetype_count.shape
```

```python
#不同户型种类的数量升序
housetype_count.sort_values()
```

```python
# 数据筛选-query()-dataframe
housetype_count_df = pd.DataFrame(data={"数量":housetype_count})
housetype_count_df = housetype_count_df.query('数量>45')
housetype_count_df
```

```python
# 绘图显示
housetype_count_df.sort_values(by=['数量'],ascending=False).plot(kind='bar',figsize=(10,4))
# 设置标题
plt.title("户型数量表")
plt.xlabel("户型")
plt.ylabel("户型数量")
# 显示图例说明
plt.legend(loc='best')
# 保存
plt.savefig('./户型数量表.png')
# 展示
plt.show()
```

```python
# 增加位置一列
file_data['位置']='北京市'+file_data['区域'].values + '区' + file_data['小区名称'].values
file_data
```

## 各个区域平均租金分析和各个区域的数量分析

```python
file_data.groupby("区域").count().shape
```

```python
# 新建dataframe对象
df_all = pd.DataFrame({"区域":file_data["区域"].unique(),
                       "房屋总金额(元)":[0]*13,
                       "房屋总面积(㎡)":[0]*13})
df_all
```

```python
# 获取每月价格--区域--总
sum_price = file_data["价格(元/月)"].groupby(file_data["区域"]).sum()
sum_price
```

```python
sum_area = file_data["面积(㎡)"].groupby(file_data["区域"]).sum()
sum_area
```

```python
df_all["房屋总金额(元)"] =  sum_price.values
df_all[ "房屋总面积(㎡)"] = sum_area.values
```

```python
df_all
```

```python
df_all["每平米租金(元)"] = round(df_all["房屋总金额(元)"]/df_all["房屋总面积(㎡)"],2)
df_all
```

```python
df_all["数量"] = house_count.values
df_all
```

```python
import matplotlib.ticker as mtick
num = df_all["数量"]
price = df_all["每平米租金(元)"]
# x轴上数据，13个区域
l = [i for i in range(13)]  
# x轴的描述信息
lx = df_all["区域"] 
```

```python
# 创建画布对象
fig = plt.figure(figsize=(10,4),dpi=100)
# 获取子图对象
ax1 = fig.add_subplot(111)
ax1.plot( l ,price,"or-",label="价格")
# y轴限制
ax1.set_ylim([0,200])
ax1.set_xlabel("房源位置")
# 显示图例说明
plt.legend(loc='upper left')

# 绘制柱状图
ax2 = ax1.twinx()  # 共用X轴
plt.bar(l, num, alpha=0.5, color='g', label='数量')
# y轴限制
ax2.set_ylim([0,2000])
# 显示图例说明
plt.legend(loc = 'upper right')

# 设置x轴
plt.xticks(l, lx)
# 保存
plt.title("价格数量表")
plt.savefig('./价格数量表.png')
plt.show()
```

# 面积区间分析
- pd.cut 离散化数据的基本使用
- 饼图的基本绘制

```python
# 获取面积数据信息
file_data["面积(㎡)"].describe()
```

```python
# 获取价格数据信息
file_data["价格(元/月)"].describe()
```

```python
# 面积数据的离散化处理
area_divide = [1, 30, 50, 70, 90, 120, 140, 160, 1200]
area_cut = pd.cut(file_data["面积(㎡)"], area_divide)
# area_cut
```

```python
# 各屋面积区间的房屋数量
ret = area_cut.value_counts().sort_index()
ret
```

```python
# 饼状图的绘制
labels = ['30平米以下', '30-50平米', '50-70平米', '70-90平米','90-120平米','120-140平米','140-160平米','160平米以上']
plt.figure(figsize=(10,10),dpi=100)
plt.pie(x=ret.values, labels=labels, autopct="%.2f%%")
plt.legend(loc='upper right')
# 保存
plt.title("房屋面积占比表")
plt.savefig('./房屋面积占比表.png')
plt.show()
```
